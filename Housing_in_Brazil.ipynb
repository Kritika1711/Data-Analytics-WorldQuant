```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df1 =pd.read_csv("brasil-real-estate-1.csv")
df1.head()
```

```python
df1.dropna(inplace=True)
df1.head()
```

```python
df1[["lat","lon"]]=df1["lat-lon"].str.split(",",expand=True).astype(float)
df1.dropna(inplace=True)
df1.shape
```

```python
df1["state"] = df1["place_with_parent_names"].str.split("|",expand=True)[2]
df1.head()
```

```python
df1["price_usd"] = df1["price_usd"].str.replace("$","", regex=False).str.replace(",","").astype(float)
```

```python
df1.info()
```

```python
df1 = df1.drop("place_with_parent_names",axis="columns")
df1.head()
```

```python
df2 = pd.read_csv("brasil-real-estate-2.csv")
```

```python
df2.head()
```

```python
df2.info()
df2.head()
```

```python
df2["price_usd"] = (df2["price_brl"]/3.19).round(2)
df2.head()
```

```python
df2 = df2.drop("price_brl",axis="columns")
df2.dropna(inplace=True)
df2.info()
```

```python
df1.info()
df1.dropna(inplace=True)

df1.info()
```

```python
df = pd.concat([df1,df2])
# df.info()
df=df.drop("lat-lon",axis="columns")
df.info()
```

```python
df.head()
```

```python
summary_stats = df[["area_m2","price_usd"]].describe()
summary_stats
```

```python
# Build histogram
plt.hist(df["price_usd"])


# Label axes
plt.xlabel("Price [USD]")
plt.ylabel("Frequency")
# Add title
plt.title("Distribution of Home Prices")
;
```

```python
# Build box plot
plt.boxplot(df["area_m2"], vert=False)


# Label x-axis
plt.xlabel("Area [sq meters]")


# Add title
plt.title("Distribution of Home Sizes")
;
```

```python
mean_price_by_region = df.groupby("region")["price_usd"].mean().sort_values(ascending=True)
mean_price_by_region
```

```python
mean_price_by_region.plot(
    kind="bar",
    xlabel="Region",
    ylabel="Mean Price[USD]",
    title="Mean Home Price by Region"
)
```
